```python
import re

def parse_model_output(model_output: str):
    result = {
        "Rating": {},
        "Rationale": {}
    }

    rating_pattern = r"Ideas:\s*(\d+).*?Organization:\s*(\d+).*?Style:\s*(\d+).*?Conventions:\s*(\d+)"
    rating_match = re.search(rating_pattern, model_output, re.DOTALL)

    if rating_match:
        result["Rating"]["Ideas"] = int(rating_match.group(1))
        result["Rating"]["Organization"] = int(rating_match.group(2))
        result["Rating"]["Style"] = int(rating_match.group(3))
        result["Rating"]["Conventions"] = int(rating_match.group(4))

    rationale_pattern = r"Ideas:\s*(.*?)\nOrganization:\s*(.*?)\nStyle:\s*(.*?)\nConventions:\s*(.*)"
    rationale_match = re.search(rationale_pattern, model_output, re.DOTALL)

    if rationale_match:
        rationale_lines = rationale_match.group(0).splitlines()
        clean = {}
        for line in rationale_lines:
            if line.startswith("Ideas:"):
                clean["Ideas"] = line.replace("Ideas:", "").strip()
            elif line.startswith("Organization:"):
                clean["Organization"] = line.replace("Organization:", "").strip()
            elif line.startswith("Style:"):
                clean["Style"] = line.replace("Style:", "").strip()
            elif line.startswith("Conventions:"):
                clean["Conventions"] = line.replace("Conventions:", "").strip()
                break  # نتوقف عند أول Conventions

        result["Rationale"] = clean
    return result
```

```python
from sklearn.metrics import cohen_kappa_score

dimensions = ["Ideas", "Organization", "Style", "Conventions"]

qwk_results = {}

for dim in dimensions:
    y_true = []
    y_pred = []

    for item in results:
        essay_id = item["essay_id"]

        # الدرجة المتوقعة من النموذج
        pred_score = item["parsed_output"]["Rating"].get(dim)

        # تجاهل القيم غير الصالحة
        if pred_score is None:
            continue

        # الدرجة الحقيقية من الداتا
        row = train_set7[train_set7["essay_id"] == essay_id].iloc[0]
        gold_score = round(
            (row["rater1_domain1"] + row["rater2_domain1"]) / 2
        )

        y_true.append(gold_score)
        y_pred.append(pred_score)

    # حساب QWK
    qwk = cohen_kappa_score(y_true, y_pred, weights="quadratic")
    qwk_results[dim] = qwk

    print(f"{dim} QWK: {qwk:.4f}")
```
